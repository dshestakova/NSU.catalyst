```python
import pandas as pd
import numpy as np

import itertools

import networkx as nx
import numpy.random as rnd
# import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
#Добавляем путь до таблички 
people = pd.read_csv('/Users/dashashestakova/Downloads/Копия Тренинг «Командообразование .catalyst»  (Ответы) - Ответы на форму (1) (копия)-4.csv')
```

```python
people.columns
```

```python
#Заполняем пустые ячейки 
people = people.fillna(-1)
people = people.replace(r'^\s+$', -1, regex=True)
```

```python
#Исправляем косяк с номерами
people = people.replace('113А', 666)
people = people.replace('113a', 666)
people = people.replace('113а', 666)
people = people.replace('113A', 666)
people = people.replace('-', -1)
```

```python
#Приводим к численному типу

people['game_1_1'] = people['game_1_1'].astype(int)
people['game_1_2'] = people['game_1_2'].astype(int)
people['game_1_3'] = people['game_1_3'].astype(int)
people['game_2_1'] = people['game_2_1'].astype(int)
people['game_2_2'] = people['game_2_2'].astype(int)
people['game_2_3'] = people['game_2_3'].astype(int)
people['game_3_1'] = people['game_3_1'].astype(int)
people['game_3_2'] = people['game_3_2'].astype(int)
people['game_3_3'] = people['game_3_3'].astype(int)
people['game_4_1'] = people['game_4_1'].astype(int)
people['game_4_2'] = people['game_4_2'].astype(int)
people['game_4_3'] = people['game_4_3'].astype(int)
people['game_5_1'] = people['game_5_1'].astype(int)
people['game_5_2'] = people['game_5_2'].astype(int)
people['game_5_3'] = people['game_5_3'].astype(int)


people['num1'] = people['num1'].astype(int)
people['num2'] = people['num2'].astype(int)
people['num3'] = people['num3'].astype(int)
people['num4'] = people['num4'].astype(int)
people['num5'] = people['num5'].fillna(-1)
people['num6'] = people['num6'].astype(int)
people['num7'] = people['num7'].astype(int)
people['num8'] = people['num8'].astype(int)
people['num9'] = people['num9'].astype(int)
people['num11'] = people['num11'].astype(int)
```

```python
num2name = people[['num', 'name']].set_index('num').to_dict()['name']
```

```python
people_dict = {}
for idx, row in people.iterrows():
    people_dict[row['name']] = {'role': row['role'], 
                                'num': str(row['num']), 
                                'connections': [str(row['num1']), str(row['num2']), str(row['num3']), str(row['num4']), str(row['num5']), str(row['num6']), str(row['num7']), str(row['num8']), str(row['num9']), str(row['num11'])],
                                'connections_game': [str(row['game_1_1']), str(row['game_1_2']), str(row['game_1_3']), str(row['game_2_1']), str(row['game_2_2']), str(row['game_2_3']), str(row['game_3_1']), str(row['game_3_2']), str(row['game_3_3']), str(row['game_4_1']), str(row['game_4_2']), str(row['game_4_3']), str(row['game_5_1']), str(row['game_5_2']), str(row['game_5_3'])]
                               }
```

```python
people_dict
```

```python
for i in people_dict.keys():
    people_dict[i]['connects_game'] = [num2name[j] if j in num2name else 0 for j in people_dict[i]['connections_game']]
    people_dict[i]['connects'] = [num2name[j] if j in num2name else 0 for j in people_dict[i]['connections']]
```

```python
for i in people_dict.keys():
    mass = []
    for j in people_dict[i]['connects']:
        if j in people_dict:
            if i in people_dict[j]['connects']:
                mass.append(1)
            else:
                mass.append(0)
        else:
            mass.append('huy')
    people_dict[i]['pairs'] = mass
```

```python
elements = []
for i in people_dict.keys():
    if people_dict[i]['role'] == 'Hustler':
        elements.append({'data': {'id': people_dict[i]['num'], 'label': i}, 'classes': 'colorectangle'})
    elif people_dict[i]['role'] == 'Hacker':
        elements.append({'data': {'id': people_dict[i]['num'], 'label': i}, 'classes': 'triangle'})
    else:
        elements.append({'data': {'id': people_dict[i]['num'], 'label': i}})

```
